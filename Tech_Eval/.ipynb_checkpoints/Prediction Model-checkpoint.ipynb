```python
# Load dataset into memory
block_data = pd.read_csv('blockchain_data.csv')
market_data = pd.read_csv('market_data.csv')
```

```python
market_data
```

```python
# Fix epoch_ts column to be pandas datetime objects and assign to index
market_data.index = [pd.to_datetime(x[:10]) for x in market_data['epoch_ts']]
block_data.index = [pd.to_datetime(x[:10]) for x in block_data['epoch_ts']]

# Sort data by ascending date
market_data = market_data.sort_index()
block_data = block_data.sort_index()
```

```python
market_data
```

```python
# Drop string columns
market_data = market_data.drop(['quote_asset_id','epoch_ts'], axis=1)
```

```python
# Separate data by underlying asset and store in dictionaries
block_dict = {}
for i in set(block_data['asset_id']):
    mask = np.where(block_data['asset_id'] == i)[0]
    block_dict[i] = block_data.iloc[mask,1:]
    
market_dict = {}
for i in set(market_data['base_asset_id']):
    mask = np.where(market_data['base_asset_id'] == i)[0]
    market_dict[i] = market_data.iloc[mask,1:]
```

```python
market_dict['Bitcoin_BTC_BTC']
```

```python
for i in market_dict['Litecoin_LTC_LTC'].iloc[:,3:]:
    #print(i)
    market_dict['Litecoin_LTC_LTC'][i].plot()
    plt.title(i)
    plt.show()
```

```python
market_dict['Bitcoin_BTC_BTC'].loc[:,'Target'] = market_dict['Bitcoin_BTC_BTC'].loc[:,'price_close'].shift(1)
```

```python
market_dict['Bitcoin_BTC_BTC'] = market_dict['Bitcoin_BTC_BTC'].dropna()
```

```python
# Segment time series for RNN model
def windowed(df,t_steps=5):
    df_arr = np.array(df.copy())
    slices = []
    for i in range(t_steps, len(df_arr)):
        slices.append(df_arr[i-t_steps:i,:])#upper bound in slice excluded
    slices = np.array(slices)
    return(slices)
Xy_slices = windowed(market_dict['Bitcoin_BTC_BTC'])
```

```python
Xy_slices.shape
```

```python
# Split into train/dev/test
dataset = Xy_slices
train = int(len(dataset) * .75)
dev = int((len(dataset) - train) * .5 + train)

X_train = dataset[:train,:,:-1]
y_train = dataset[:train,-1,-1]
X_dev = dataset[train:dev,:,:-1]
y_dev = dataset[train:dev,-1,-1]
X_test = dataset[dev:,:,:-1]
y_test = dataset[dev:,-1,-1]
```

```python
# Scale data
import sys
eps = sys.float_info.epsilon
mu = X_train.mean()
s = X_train.std()

X_train -= mu
X_train = (X_train + eps) / s

X_dev -= mu
X_dev = (X_dev + eps) / s

X_test -= mu
X_test = (X_test + eps) / s
```

```python
print(X_train.shape,y_train.shape)
```

```python
# Import the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout

# Initialising the RNN
regressor = Sequential()

# Adding first LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 50, 
                   return_sequences = True, 
                   input_shape = (X_train.shape[1], X_train.shape[2])
                   ))
regressor.add(Dropout(0.2))

# Adding additional LSTM layers with Dropout regularisation
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = False))
regressor.add(Dropout(0.2))

# Final Layer
regressor.add(Dense(units = 1))

# Compiling the RNN
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

# Fitting the RNN to the training set
regressor.fit(X_train, y_train, epochs = 10,  batch_size = 8064)


# Making Predictions and visualising the results
#dataset_total = pd.concat((pd.DataFrame(train_data), pd.DataFrame(test_data)), axis = 0)
```
